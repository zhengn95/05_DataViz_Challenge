### Analysis
This analysis uses complete data from Pymaceuticals's most recent animal study. The purpose of this study was to compare the performance of Pymaceuticalsâ€™ drug of interest, Capomulin, against the other treatment regimens.

In this analysis, I was tasked with generating all of the tables and figures needed for the technical report of the clinical study and a summary of the study results (see README.md file).


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, how = 'right', on = ['Mouse ID', 'Mouse ID'])
reordered_df = ['Mouse ID', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Drug Regimen', 
    'Sex', 'Age_months', 'Weight (g)']
reordered_merged_df = merged_df[reordered_df]

# Display the data table for preview
reordered_merged_df.head()
```

```python
# Checking the number of mice.
len(reordered_merged_df['Mouse ID'].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_df = reordered_merged_df[reordered_merged_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicated_df['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicated_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = reordered_merged_df[reordered_merged_df['Mouse ID'] != 'g989']
cleaned_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
groupby_drug = cleaned_df.groupby('Drug Regimen')

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_df = groupby_drug['Tumor Volume (mm3)'].mean()
median_df = groupby_drug['Tumor Volume (mm3)'].median()
variance_df = groupby_drug['Tumor Volume (mm3)'].var()
std_df = groupby_drug['Tumor Volume (mm3)'].std()
sem_df = groupby_drug['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.
SummaryStats_df = pd.DataFrame({'Mean Tumor Volume': mean_df, 'Median Tumor Volume': median_df, 
                               'Tumor Volume Variance': variance_df, 'Tumor Volume Std. Dev.': std_df,
                               'Tumor Volume Std. Err.': sem_df})
SummaryStats_df.sort_values('Drug Regimen')
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
SummaryStats_agg = groupby_drug.agg({'Tumor Volume (mm3)' : ['mean', 'median', 'var', 'std', 'sem']})
SummaryStats_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
grouped_mouse = cleaned_df.groupby(['Mouse ID'])['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
grouped_merged = pd.merge(cleaned_df, grouped_mouse, on = ['Mouse ID', 'Timepoint'], how = 'right')
grouped_merged
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
        
# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = {}

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in regimens:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    subset = grouped_merged.loc[grouped_merged['Drug Regimen'] == regimen, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_volumes[regimen] = subset.tolist()
    
    # Determine outliers using upper and lower bounds
    Q1 = subset.quantile(0.25)
    Q3 = subset.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound  = Q3 + (1.5 * IQR)
    outliers = subset.loc[(subset < lower_bound) | (subset > upper_bound)]

final_tumor_volume = [tumor_volumes[regimen] for regimen in regimens]
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
plt.boxplot(final_tumor_volume, labels=regimens,flierprops=dict(markerfacecolor='red', marker='o', markersize=12))
plt.title('Tumor Volume Distribution by Treatment Group')
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```
